```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

```python
# Load dataset
df = pd.read_csv("dataset.csv")
df.head()
```

```python
# Check dataset info
print(df.shape)  # number of rows and columns
df.info()        # data types and missing values
df.describe()    # basic statistics of numeric columns
```

```python
# Check missing values
print(df.isnull().sum())
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.show()
```

```python
# Handle missing values (only Cuisines has missing values)
df['Cuisines'] = df['Cuisines'].fillna('Unknown')
```

```python
# Check data type of Aggregate rating
print(df['Aggregate rating'].dtype)
```

```python
# Data Type Check
# Aggregate rating is already numeric (float64), so no conversion needed.
```

```python
# Analyze distribution
print(df['Aggregate rating'].value_counts())

import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(df['Aggregate rating'], bins=20)
plt.show()
```

```python
# Check class imbalance
sns.countplot(x='Aggregate rating', data=df)
plt.show()
```

### Task 1 Summary: Data Exploration and Preprocessing

1. **Dataset Overview:** The dataset contains 9,551 rows and 21 columns.
2. **Missing Values:** Only the `Cuisines` column had missing values, which were filled with `'Unknown'`.
3. **Data Types:** `Aggregate rating` is numeric (`float64`), so no conversion was needed.
4. **Target Variable Analysis:** The distribution of `Aggregate rating` was examined using value counts and a histogram.
5. **Class Imbalance:** Checked using a count plot; ratings are slightly imbalanced.
6. **Conclusion:** The dataset is now clean, properly formatted, and ready for further analysis.

```python
# Task 3: Geospatial Analysis
# Objective: Visualize restaurant locations on a map, analyze distribution across cities, and explore any correlation between location and restaurant ratings.
# Code and Analysis
```

```python
# Scatter plot: Restaurant locations colored by aggregate rating
plt.figure(figsize=(5,3))
plt.scatter(df['Longitude'],df['Latitude'], c=df['Aggregate rating'], cmap='viridis', alpha=0.5)
plt.colorbar(label='Aggregate Rating')  # Color indicates rating
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Restaurant Locations by Rating')
plt.show()
```

### Explanation

- `df['Longitude']` & `df['Latitude']`: Place each restaurant on the map.
- `c=df['Aggregate rating']`: Colors points according to rating, allowing visual comparison.
- `cmap='viridis'`: Uses a professional color gradient for rating values.
- `alpha=0.5`: Makes points semi-transparent, preventing overlap from hiding data.
- `plt.colorbar()`: Adds a legend to interpret colors.
- Axis labels & title: Provide context and clarity for the plot.


```python
# 2. Bar plot: Top 10 cities by number of restaurants
top_cities = df['City'].value_counts().head(10)
top_cities.plot(kind='bar', figsize=(5,3), color='skyblue', title='Top 10 Cities by Number of Restaurants')
plt.ylabel('Number of Restaurants')
plt.show()
```

```python
# 3. Bar plot: Top 10 cities by average restaurant rating
top_avg_rating_cities = df.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False).head(10)
top_avg_rating_cities.plot(kind='bar', figsize=(5,3), color='green', title='Top 10 Cities by Average Rating')
plt.ylabel('Average Rating')
plt.show()
```

```python
# 4. Correlation between coordinates and rating
coord_corr = df[['Latitude', 'Longitude', 'Aggregate rating']].corr()
print("Correlation between Latitude/Longitude and Aggregate Rating:\n", coord_corr)
```

### Summary: Geospatial Analysis

- Plotted restaurant locations on a scatter plot using latitude and longitude, with color indicating aggregate rating.
- Identified the top 10 cities with the highest number of restaurants.
- Analyzed the top 10 cities based on average restaurant ratings.
- Calculated correlation between restaurant location (latitude and longitude) and aggregate rating:
  - Latitude vs Rating: 0.0005  No correlation
  - Longitude vs Rating: -0.1168  Very weak negative correlation
- Conclusion: Restaurant ratings are largely independent of their geographical location.
